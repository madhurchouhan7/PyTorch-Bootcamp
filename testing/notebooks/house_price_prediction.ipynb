```python
label_encoder = LabelEncoder()
```

```python
df['mainroad'] = label_encoder.fit_transform(df['mainroad'])
df['guestroom'] = label_encoder.fit_transform(df['guestroom'])
df['basement'] = label_encoder.fit_transform(df['basement'])
df['airconditioning'] = label_encoder.fit_transform(df['airconditioning'])
df['prefarea'] = label_encoder.fit_transform(df['prefarea'])
df['furnishingstatus'] = label_encoder.fit_transform(df['furnishingstatus'])
df['hotwaterheating'] = label_encoder.fit_transform(df['hotwaterheating'])
```

```python
X_raw = df.drop('price', axis=1)
y_raw = df['price']
```

```python
y_raw
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_raw)
```

```python
target_scaler = StandardScaler()

Y_scaled = target_scaler.fit_transform(y_raw.values.reshape(-1, 1))
```

```python
X_scaled
```

```python
Y_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y_scaled, test_size=0.2, random_state=42)
```

```python
class HousingDataset(Dataset):
    def __init__(self, features, targets):
        self.X = torch.tensor(features, dtype=torch.float32)
        self.y = torch.tensor(targets, dtype=torch.float32).view(-1, 1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_housing_dataset = HousingDataset(X_train, y_train)

test_housing_dataset = HousingDataset(X_test, y_test)
```

```python
train_loader = DataLoader(train_housing_dataset, batch_size=32, shuffle=True)

for batch_X, batch_y in train_loader:
    print("Batch X Shape:", batch_X.shape)
    print("Batch y Shape:", batch_y.shape)
    break
```

```python
test_loader = DataLoader(test_housing_dataset, batch_size=32, shuffle=False)
```

```python
class HousePricePredictor(nn.Module):
    def __init__(self, input_dim):
        super(HousePricePredictor, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
model = HousePricePredictor(input_dim=X_scaled.shape[1])
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 50

for epoch in range(epochs):

    epoch_loss = 0

    for batch_X, batch_y in train_loader:

        predictions = model(batch_X)

        loss = criterion(predictions, batch_y)

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        epoch_loss += loss.item() * batch_X.size(0)

    epoch_loss /= len(train_loader.dataset)

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}")
```

```python
model.eval()

with torch.no_grad():
    test_loss = 0

    for batch_X, batch_y in test_loader:

        predictions = model(batch_X)

        loss = criterion(predictions, batch_y)

        test_loss += loss.item() * batch_X.size(0)

    test_loss /= len(test_loader.dataset)

    print(f"Test Loss: {test_loss:.4f}")
```

```python
torch.save(model.state_dict(), "model/house_price_predictor.pth")
```
